```python
import boto3
import os
import numpy as np 

from sagemaker.session import Session

from PyPDF2 import PdfFileReader, PdfFileWriter, utils
```

```python
bucket = "ran-s3-systemic-risk"
data_folder = 'Input/X-17A-5/'
export_folder = "Input/SubSets/"

# Amazon Textract client and Sagemaker session
textract = boto3.client('textract')
s3 = boto3.client('s3')
session = Session()
```

```python
importPaths = np.array(session.list_s3_files(bucket, data_folder))
```

```python
importPaths
```

```python
# pages to keep from each pdf
pages = np.arange(2,11) # page 1, 3, 5

# export file paths to document subfolder
exportPaths = np.array(session.list_s3_files(bucket, export_folder))

for pdf_file_path in importPaths[1:]:
    
    # check to see if values are downloaded to s3 sub-bin
    baseFile = pdf_file_path.split('/')[-1].split('.')[0]
    exportFile = '{0}_subset.pdf'.format(baseFile)
    
    if export_folder + exportFile not in exportPaths:
        
        try:
            # retrieving downloaded files from s3 bucket
            s3.download_file(bucket, pdf_file_path, 'temp.pdf')

            # read pdf file and initialize empty pdf file
            pdf = PdfFileReader('temp.pdf')
            pdfWriter = PdfFileWriter()
            
            # since some files are smaller than 11 pages
            try:
                # selectively add the pages from the document 
                for page_num in pages:
                    pdfWriter.addPage(pdf.getPage(page_num))
            except IndexError:
                pass

            # open file and save to local instance
            try:
                with open(exportFile, 'wb') as f:
                    pdfWriter.write(f)
                    f.close()
                    
                # save contents to AWS S3 bucket
                with open(exportFile, 'rb') as data:
                    s3.upload_fileobj(data, bucket, export_folder + exportFile)
                print('Saved file -> {}'.format(exportFile))
            except:
                print('{} was not able to be saved'.format(exportFile))

            # remove local file after it has been created
            os.remove('temp.pdf')
            os.remove(exportFile)

        except utils.PdfReadError:
            print('EOF marker not found - reject')
    
    else:
        print('{} already saved'.format(exportFile))
        
```

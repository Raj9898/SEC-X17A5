```python
already_done = np.zeros(len(raw_broker_dealer_pdfs))
out_folder_raw_pdf = temp_folder + 'X-17A-5-PDF-RAW/'

bucket = "x17-a5-mathias-version-nit"
job_ids = len(raw_broker_dealer_pdfs)*[0]

for counter, pdf_paths in enumerate(textract_files):
    if counter < 46660:
        continue
    if counter > 47660:
        continue

    basefile = pdf_paths.split('/')[-1].split('-subset')[0]
    fileName = basefile + '.csv'
    if (out_folder_raw_pdf + fileName in output_pdf_csvs):
        already_done[counter] = 1
    job_ids[counter] = str(counter)
    
    png_paths = input_png + basefile + '/'

    if counter%200 == 0:
        print(counter)
    
    while True:
        try:
            job_ids[counter] = startJob(bucket, pdf_paths)
            break
        except Exception as e:
            print(e)
            time.sleep(1)

    
```
